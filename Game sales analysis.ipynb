#### Проверка гипотез

В этой главе проверим следующие гипотезы:

- Средние пользовательские рейтинги платформ *Xbox One* и *PC* одинаковые;
- Средние пользовательские рейтинги жанров *Action* (англ. «действие», экшен-игры) и *Sports* (англ. «спортивные соревнования») разные.

Для проверки этих гипотез мы можем использовать t-тест для независимых выборок. Для проверки гипотезы мы выбираем t-тест, так как он позволяет сравнить средние значения двух групп и определить, есть ли статистически значимая разница между ними. В данном случае у нас две независимые выборки (пользовательские рейтинги платформ и жанров), и мы хотим определить, различаются ли средние значения этих выборок. T-тест подходит для такого типа сравнения. Пороговое значение alpha выберем общепринятое значение 0.05.


Сформулируем нулвую и алтернитвую гипотезу для певрого утверждения:
 - Нулевая гипотеза (H0): Средние пользовательские рейтинги платформ Xbox One и PC равны.
 - Альтернативная гипотеза (H1): Средние пользовательские рейтинги платформ Xbox One и PC различаются.
